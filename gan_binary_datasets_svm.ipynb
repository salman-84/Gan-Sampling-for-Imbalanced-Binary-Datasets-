```python
count= 1

for ii in data_name:
    data= pd.read_csv(r'E:\research\datasets\binary\{}.csv'.format(ii))
    x=data.iloc[:,:-1]
    y=data.iloc[:,-1]
    ir=y.value_counts()[0]/y.value_counts()[1];
    ir=round(ir,2)

    acc= pd.concat([acc, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    gmean= pd.concat([gmean, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    sensitivity= pd.concat([sensitivity, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    specificity=pd.concat([specificity, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    roc= pd.concat([roc, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    precision= pd.concat([precision, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    mcc= pd.concat([mcc, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])
    f1score= pd.concat([f1score, pd.DataFrame({'dataset':ii, 'Imbalance Ratio':ir},index=[0])])

    
    count= count+1
    
```

```python
count= 1

for ii in data_name:
    data= pd.read_csv(r'E:\research\datasets\binary\{}.csv'.format(ii))
    x=data.iloc[:,:-1]
    y=data.iloc[:,-1]


    X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=84)

    
    svm = SVC()

    svm.fit(X_train, y_train)
    y_preds = svm.predict(X_test)
    
    gmean.loc[count-1, 'SVM']= geometric_mean_score(y_test,y_preds)
    roc.loc[count-1, 'SVM']= roc_auc_score(y_test,y_preds)
    mcc.loc[count-1, 'SVM']= matthews_corrcoef(y_test,y_preds)
    sensitivity.loc[count-1, 'SVM']= sensitivity_score(y_test,y_preds)
    specificity.loc[count-1, 'SVM']= specificity_score(y_test,y_preds)
    acc.loc[count-1, 'SVM']= accuracy_score(y_test,y_preds)
    precision.loc[count-1, 'SVM'] = precision_score(y_test,y_preds)
    f1score.loc[count-1, 'SVM']= f1_score(y_test,y_preds)


    
    count= count+1
    
```
